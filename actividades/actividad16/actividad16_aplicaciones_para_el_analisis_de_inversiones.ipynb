```python
# CLCULO 2: SEGUNDO SEMESTRE
print("\n2. ANLISIS SEGUNDO SEMESTRE")
print("="*40)

# Calcular tasa efectiva del segundo semestre (compuesta)
tasa_2do_semestre = 1
for tasa in inflacion_2do_semestre:
    tasa_2do_semestre = tasa_2do_semestre * (1 + tasa)
tasa_2do_semestre = tasa_2do_semestre - 1

# Calcular tasa simple del segundo semestre
tasa_2do_semestre_simple = sum(inflacion_2do_semestre)

# Anualizar el segundo semestre usando la funci贸n importada (m=2 porque hay 2 semestres por a帽o)
tasa_2do_semestre_anualizada = annualized_rate(tasa_2do_semestre, 2)

print(f"Tasa semestral simple (suma): {tasa_2do_semestre_simple*100:.2f}%")
print(f"Tasa semestral efectiva (compuesta): {tasa_2do_semestre*100:.2f}%")
print(f"Tasa anualizada (proyecci贸n): {tasa_2do_semestre_anualizada*100:.2f}%")
print(f"Diferencia (compuesta vs simple): {(tasa_2do_semestre - tasa_2do_semestre_simple)*100:.2f}%")
```

```python
# CLCULO 3: INFLACIN ANUAL NOMINAL
print("\n3. ANLISIS ANUAL NOMINAL")
print("="*40)

# Combinar ambos semestres
inflacion_anual_completa = inflacion_1er_semestre + inflacion_2do_semestre

# M茅todo 1: Inflaci贸n anual simple (suma aritm茅tica)
inflacion_anual_simple = sum(inflacion_anual_completa)

# M茅todo 2: Inflaci贸n anual compuesta (capitalizaci贸n mensual)
inflacion_anual_compuesta = 1
for tasa_mensual in inflacion_anual_completa:
    inflacion_anual_compuesta = inflacion_anual_compuesta * (1 + tasa_mensual)
inflacion_anual_compuesta = inflacion_anual_compuesta - 1

# Comparaci贸n de m茅todos
diferencia_metodos = inflacion_anual_compuesta - inflacion_anual_simple

print(f"Inflaci贸n anual simple (suma): {inflacion_anual_simple*100:.2f}%")
print(f"Inflaci贸n anual compuesta (capitalizada): {inflacion_anual_compuesta*100:.2f}%")
print(f"Diferencia entre m茅todos: {diferencia_metodos*100:.2f}%")

print("\nRESUMEN FINAL:")
print("="*40)
print(f" Primer semestre anualizado: {tasa_1er_semestre_anualizada*100:.2f}%")
print(f" Segundo semestre anualizado: {tasa_2do_semestre_anualizada*100:.2f}%")
print(f" Inflaci贸n anual real (compuesta): {inflacion_anual_compuesta*100:.2f}%")
```

### Ejercicio 2: Proporci贸n

Tienes un capital de $1000 que quieres invertir. Un banco te ofrece una Tasa Nominal Anual (TNA) del 48% con capitalizaci贸n mensual.

1.  Calcula la Tasa Efectiva Mensual (TEM)
2.  Calcula la Tasa Efectiva Anual (TEA)


```python
# EJERCICIO 2: C谩lculo de Tasas Efectivas usando funciones de matem谩ticas financieras
from utils.matematicas_financieras import nominal_to_effective_rate, effective_annual_rate

# Datos del problema
capital = 1000  # Capital inicial en pesos
tna = 0.48      # Tasa Nominal Anual del 48%
m = 12          # Per铆odos de capitalizaci贸n por a帽o (mensual)

print("=== EJERCICIO 2: CLCULO DE TASAS EFECTIVAS ===")
print(f"Capital inicial: ${capital}")
print(f"TNA: {tna*100}%")
print(f"Capitalizaci贸n: mensual ({m} veces por a帽o)")
print()

# 1. C谩lculo de la Tasa Efectiva Mensual (TEM)
print("1. CLCULO DE LA TASA EFECTIVA MENSUAL (TEM)")
tem = nominal_to_effective_rate(tna, m)
print(f"Usando funci贸n: nominal_to_effective_rate({tna}, {m})")
print(f"TEM = {tem:.4f} = {tem*100:.2f}%")
print()

# 2. C谩lculo de la Tasa Efectiva Anual (TEA)
print("2. CLCULO DE LA TASA EFECTIVA ANUAL (TEA)")
tea = effective_annual_rate(tna, m)
print(f"Usando funci贸n: effective_annual_rate({tna}, {m})")
print(f"TEA = {tea:.6f} = {tea*100:.2f}%")
print()

# Resumen y demostraci贸n pr谩ctica
print("=== RESUMEN DE RESULTADOS ===")
print(f" Tasa Nominal Anual (TNA): {tna*100:.2f}%")
print(f" Tasa Efectiva Mensual (TEM): {tem*100:.2f}%")
print(f" Tasa Efectiva Anual (TEA): {tea*100:.2f}%")
print()

print("=== DEMOSTRACIN PRCTICA ===")
monto_final = capital * (1 + tea)
print(f"Invirtiendo ${capital} por un a帽o al {tea*100:.2f}% TEA:")
print(f"Monto final: ${monto_final:.2f}")
print(f"Ganancia: ${monto_final - capital:.2f}")
```

### Ejercicio 3: Valor Actual Neto (VAN)

Una empresa est谩 evaluando un proyecto de inversi贸n que requiere una inversi贸n inicial de $50,000 en el A帽o 0. Se esperan los siguientes flujos de caja en los pr贸ximos 4 a帽os:

*   A帽o 1: $15,000

*   A帽o 2: $20,000

*   A帽o 3: $25,000

*   A帽o 4: $10,000

Si la tasa de descuento requerida (costo de capital) es del 10% anual, calcula el Valor Actual Neto (VAN) del proyecto e interpreta si el proyecto es rentable.

```python
# EJERCICIO 3: An谩lisis completo de inversi贸n (VAN, TIR y Per铆odo de Recuperaci贸n)
from utils.matematicas_financieras import (net_present_value, present_value, 
                                          internal_rate_of_return, payback_period)

# Datos del proyecto
inversion_inicial = 50000
flujos_futuros = [15000, 20000, 25000, 10000]
tasa_descuento = 0.10
tiempo_max_recuperacion = 3.0

print("=== ANLISIS DE EVALUACIN DE INVERSIN ===")
print(f"Inversi贸n inicial: ${inversion_inicial:,.2f}")
print(f"Flujos de caja: {[f'${f:,.0f}' for f in flujos_futuros]}")
print(f"Tasa de descuento: {tasa_descuento*100:.1f}%\n")

# 1. C谩lculo del VAN
van = net_present_value(inversion_inicial, flujos_futuros, tasa_descuento)
print(f"1. VAN = ${van:,.2f}")

# 2. C谩lculo de la TIR
tir = internal_rate_of_return(inversion_inicial, flujos_futuros)
print(f"2. TIR = {tir*100:.2f}%")

# 3. Per铆odo de Recuperaci贸n Simple
pr_simple = payback_period(inversion_inicial, flujos_futuros)
print(f"3. Per铆odo de Recuperaci贸n Simple = {pr_simple:.2f} a帽os")

# 4. Per铆odo de Recuperaci贸n Descontado
valores_presentes = [present_value(flujo, tasa_descuento, i+1, 1) 
                    for i, flujo in enumerate(flujos_futuros)]
pr_descontado = payback_period(inversion_inicial, valores_presentes)
print(f"4. Per铆odo de Recuperaci贸n Descontado = {pr_descontado:.2f} a帽os\n")

# An谩lisis de criterios
print("=== ANLISIS DE RENTABILIDAD ===")
criterios = {
    'VAN positivo': van > 0,
    'TIR > tasa descuento': tir > tasa_descuento,
    'Recuperaci贸n aceptable': pr_descontado <= tiempo_max_recuperacion
}

for criterio, cumple in criterios.items():
    estado = "S" if cumple else "NO"
    print(f" {criterio}: {estado}")

# Recomendaci贸n
criterios_positivos = sum(criterios.values())
print(f"\nCriterios favorables: {criterios_positivos}/3")

if criterios_positivos >= 2:
    print("RECOMENDACIN: ACEPTAR EL PROYECTO")
else:
    print("RECOMENDACIN: EVALUAR CON CAUTELA O RECHAZAR")

# Resumen econ贸mico
if van > 0:
    rendimiento_por_peso = 1 + (van/inversion_inicial)
    print(f"\nCada peso invertido genera ${rendimiento_por_peso:.3f}")
```

#### An谩lisis de Riesgo con Simulaci贸n Monte Carlo

Para complementar el an谩lisis tradicional, realizaremos una **simulaci贸n Monte Carlo** que considera la incertidumbre en los flujos de caja y la tasa de descuento. Esto nos permitir谩:

-  **Cuantificar el riesgo** del proyecto bajo diferentes escenarios
-  **Calcular la probabilidad de 茅xito** (NPV  0)
-  **Visualizar la distribuci贸n** de resultados posibles
-  **Tomar decisiones m谩s informadas** considerando la incertidumbre

**Supuestos de la simulaci贸n:**
- **Volatilidad de flujos**: 卤20% (variabilidad t铆pica en proyectos empresariales)
- **Volatilidad de tasa**: 卤2% (incertidumbre en costo de capital)
- **Simulaciones**: 1,000 escenarios diferentes

```python
# SIMULACIN MONTE CARLO para an谩lisis de riesgo
from utils.matematicas_financieras import monte_carlo_npv, plot_monte_carlo_results

print(" INICIANDO SIMULACIN MONTE CARLO...")
print("=" * 50)

# Preparar datos para Monte Carlo
# Los flujos incluyen la inversi贸n inicial como primer elemento (negativo)
cash_flows_mc = [-inversion_inicial] + flujos_futuros
print(f"Flujos de caja para simulaci贸n: {[f'${f:,.0f}' for f in cash_flows_mc]}")

# Par谩metros de incertidumbre
volatilidad_flujos = 0.20    # 卤20% de variabilidad en flujos de caja
volatilidad_tasa = 0.02      # 卤2% de variabilidad en tasa de descuento
num_simulaciones = 100000    # N煤mero de escenarios a simular

print(f" Tasa de descuento base: {tasa_descuento:.1%}")
print(f" Volatilidad de flujos: 卤{volatilidad_flujos:.1%}")
print(f" Volatilidad de tasa: 卤{volatilidad_tasa:.1%}")
print(f" Simulaciones: {num_simulaciones:,}")
print()

# Ejecutar simulaci贸n Monte Carlo
resultados_mc = monte_carlo_npv(
    cash_flows=cash_flows_mc,
    discount_rate_mean=tasa_descuento,
    volatility_flows=volatilidad_flujos,
    volatility_rate=volatilidad_tasa,
    simulations=num_simulaciones
)

print(f" Simulaci贸n completada: {len(resultados_mc):,} escenarios generados")
print()

# Generar visualizaci贸n y obtener m茅tricas
print(" GENERANDO ANLISIS VISUAL Y ESTADSTICO...")
metricas_mc = plot_monte_carlo_results(
    resultados_mc, 
    "Ejercicio 3: An谩lisis de Riesgo del Proyecto de Inversi贸n"
)

# Mostrar resultados comparativos
print("=" * 50)
print(" COMPARACIN: ANLISIS DETERMINSTICO vs MONTE CARLO")
print("=" * 50)
print(f"VAN Determin铆stico:        ${van:,.2f}")
print(f"VAN Promedio Monte Carlo:  ${metricas_mc['npv_mean']:,.2f}")
print(f"Diferencia:               ${metricas_mc['npv_mean'] - van:,.2f}")
print()

print(" ANLISIS DE RIESGO")
print("-" * 30)
print(f" Riesgo (Desv. Est谩ndar):     ${metricas_mc['npv_std']:,.2f}")
print(f" Probabilidad de xito:       {metricas_mc['success_probability']:.1%}")
print(f" Probabilidad de P茅rdida:     {metricas_mc['failure_probability']:.1%}")
print()

print(" ESCENARIOS EXTREMOS (90% de confianza)")
print("-" * 30)
print(f" Escenario Pesimista (5%):    ${metricas_mc['percentile_5']:,.2f}")
print(f" Escenario Optimista (95%):   ${metricas_mc['percentile_95']:,.2f}")
print(f" Rango de Variaci贸n:          ${metricas_mc['percentile_95'] - metricas_mc['percentile_5']:,.2f}")
print()

# Interpretaci贸n y recomendaci贸n final
print(" INTERPRETACIN INTEGRADA")
print("-" * 30)

# Criterios combinados
van_mc_positivo = metricas_mc['npv_mean'] > 0
alta_probabilidad_exito = metricas_mc['success_probability'] > 0.70
riesgo_aceptable = metricas_mc['npv_std'] / abs(metricas_mc['npv_mean']) < 1.0

print("Criterios de Evaluaci贸n (con Monte Carlo):")
print(f" VAN Promedio Positivo:       {'S' if van_mc_positivo else 'NO'}")
print(f" Probabilidad xito > 70%:    {'S' if alta_probabilidad_exito else 'NO'}")
print(f" Riesgo Aceptable:            {'S' if riesgo_aceptable else 'NO'}")
print()

# Recomendaci贸n final considerando riesgo
if van_mc_positivo and alta_probabilidad_exito:
    recomendacion_final = " ACEPTAR EL PROYECTO"
    justificacion = "VAN positivo con alta probabilidad de 茅xito y riesgo controlado"
elif van_mc_positivo:
    recomendacion_final = " ACEPTAR CON PRECAUCIN"
    justificacion = "VAN positivo pero con riesgo significativo - considerar medidas de mitigaci贸n"
else:
    recomendacion_final = " RECHAZAR EL PROYECTO"
    justificacion = "VAN promedio negativo indica alta probabilidad de p茅rdida"

print(f"RECOMENDACIN FINAL: {recomendacion_final}")
print(f"JUSTIFICACIN: {justificacion}")
print("=" * 50)
```
